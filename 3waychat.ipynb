```python
import os 
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
if api_key:
    print("Yes we have the openAI api key")
else:
    print("OpenAI API key not found")
```

```python
system_prompt_1 = """
You are a creative AI assistant.

Whenever a user asks a question, think creatively.
Offer imaginative ideas and examples.
Keep your answers concise.
"""
system_prompt_2 = """
You are a critical thinker.

Read the previous assistant's answer.
Find mistakes, weaknesses, assumptions or missing details.
Suggest improvements politely.
"""
system_prompt_3 = """
You are an expert evaluator.

Read both previous responses.

Produce the final answer by:
- choosing the strongest ideas
- correcting mistakes
- combining both answers

Return only the final response.
"""
```

```python
openai = OpenAI()
def chat(model,message):
    stream = openai.chat.completions.create(model=model, messages= message, stream=True)
    response = ""

    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
messageA = [
    {
        "role": "system",
        "content": system_prompt_1
    },
    {
        "role": "user",
        "content": "Why is Python popular?"
    }
]
responseA = ""
for text in chat("gpt-5.5",messageA):
    responseA = text;
messageB = [
    {
        "role": "system",
        "content": system_prompt_2
    },
    {
        "role": "user",
        "content": "Why is Python popular?"
    },
     {
        "role": "assistant",
        "content": responseA
    }
]
responseB = ""
for text in chat("gpt-5.4-mini",messageB):
    responseB = text;


messageC = [
    {
        "role": "system",
        "content": system_prompt_3
    },
    {
        "role": "user",
        "content": "Why is Python popular?"
    },
    {
        "role": "assistant",
        "content": responseA
    },
    {
        "role": "assistant",
        "content": responseB
    }
]
responseC = ""
for text in chat("gpt-5.4-nano",messageC):
    responseC = text;
```

```python
from IPython.display import display, Markdown
for i in range(5):
    display(Markdown(f"#Round{i+1}"))
    responseA = ""
    for text in chat("gpt-5.5",messageA):
        responseA = text
    display(Markdown(f"## Model A\n\n{responseA}"))
    messageA.append({"role":"assistant","content":responseA})
    messageB.append({"role":"user","content":responseA})
    responseB = ""
    for text in chat("gpt-5.4-mini",messageB):
        responseB = text
    display(Markdown(f"## Model B\n\n{responseB}"))
    messageB.append({"role":"assistant","content":responseB})
    messageC.append({"role":"user","content":responseB})
    responseC = ""
    for text in chat("gpt-5.4-nano",messageC):
        responseC = text;
    display(Markdown(f"## Model C\n\n{responseC}"))
    messageC.append({"role":"assistant","content":responseC})
    messageA.append({"role":"user","content":responseC})

    

    


```
